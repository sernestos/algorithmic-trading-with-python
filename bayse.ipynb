Exemple d'ehantillonage du post à l'aide d'une grille


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Définir les données et paramètres du problème
n = 20   # Nombre total de lancers
x = 14   # Nombre de faces obtenues
grid_size = 1000  # Nombre de points dans la grille

# 2. Créer une grille de valeurs pour θ
theta_grid = np.linspace(0, 1, grid_size+1)

# 3. Calculer la vraisemblance non normalisée à chaque point de la grille
# Prior uniforme p(θ) = 1
likelihood = theta_grid**x * (1 - theta_grid)**(n - x)  # p(x|θ)
posterior_unnorm = likelihood  # p(θ|x) ∝ p(x|θ)p(θ) et p(θ) = 1

# 4. Normaliser pour obtenir les probabilités
normalizer = np.sum(posterior_unnorm)
posterior_probs = posterior_unnorm / normalizer

# 5. Visualiser la distribution a posteriori
plt.figure(figsize=(10, 6))
plt.plot(theta_grid, posterior_probs)
plt.title("Distribution a posteriori")
plt.xlabel("θ")
plt.ylabel("p(θ|x)")
plt.grid(True)
plt.show()

# 6. Créer une fonction de répartition empirique (CDF)
cdf = np.cumsum(posterior_probs)

# 7. Fonction pour échantillonner de la distribution a posteriori
def sample_from_posterior(num_samples):
    u = np.random.random(num_samples)
    samples = np.zeros(num_samples)

    for i in range(num_samples):
        # Trouver l'index où u[i] est inférieur à la CDF
        idx = np.searchsorted(cdf, u[i])
        samples[i] = theta_grid[idx]

    return samples

# 8. Générer 1000 échantillons et visualiser l'histogramme
posterior_samples = sample_from_posterior(1000)

plt.figure(figsize=(10, 6))
plt.hist(posterior_samples, bins=20, density=True, alpha=0.7)
plt.title("Échantillons de la distribution a posteriori")
plt.xlabel("θ")
plt.ylabel("Fréquence relative")
plt.grid(True)
plt.show()

# 9. Calculer des statistiques à partir des échantillons
print(f"Moyenne a posteriori: {np.mean(posterior_samples):.4f}")
print(f"Médiane a posteriori: {np.median(posterior_samples):.4f}")
print(f"Écart-type a posteriori: {np.std(posterior_samples):.4f}")

# 10. Intervalle de crédibilité à 95%
sorted_samples = np.sort(posterior_samples)
credible_interval = sorted_samples[[24, 974]]  # Indices pour 2.5% et 97.5%
print(f"Intervalle de crédibilité à 95%: [{credible_interval[0]:.4f}, {credible_interval[1]:.4f}]")

# 11. Comparer avec la solution analytique (loi Beta)
a = x + 1  # Paramètre alpha de la Beta
b = n - x + 1  # Paramètre beta de la Beta
exact_beta = stats.beta(a, b)

# Tracer la solution exacte
theta_values = np.linspace(0, 1, 1000)
plt.figure(figsize=(10, 6))
plt.plot(theta_values, exact_beta.pdf(theta_values), 'r-', label='Solution exacte (Beta)')
plt.plot(theta_grid, posterior_probs, 'b-', label='Approximation par grille')
plt.title("Comparaison: Grille vs. Solution exacte")
plt.xlabel("θ")
plt.ylabel("p(θ|x)")
plt.legend()
plt.grid(True)
plt.show()

# 12. Tracer la CDF empirique
plt.figure(figsize=(10, 6))
plt.plot(theta_grid, cdf, 'b-')
plt.plot(theta_values, exact_beta.cdf(theta_values), 'r--')
plt.title("Fonction de répartition empirique vs exacte")
plt.xlabel("θ")
plt.ylabel("CDF")
plt.legend(['CDF empirique', 'CDF exacte (Beta)'])
plt.grid(True)
plt.show()

# 13. Calculer la probabilité que θ > 0.5
prob_theta_gt_half_empirical = 1 - cdf[np.searchsorted(theta_grid, 0.5)]
prob_theta_gt_half_exact = 1 - exact_beta.cdf(0.5)

print(f"P(θ > 0.5) [empirique]: {prob_theta_gt_half_empirical:.4f}")
print(f"P(θ > 0.5) [exacte]: {prob_theta_gt_half_exact:.4f}")
```
